```python
# Creating new training samples by randomly cropping existing images and their masks
output_image_dir = "aug_images"
output_mask_dir = "aug_masks"
os.makedirs(output_image_dir, exist_ok=True)
os.makedirs(output_mask_dir, exist_ok=True)


def clean_folder(folder):
    files = glob(os.path.join(folder, "*"))

    for f in files:
        if os.path.isfile(f):
            os.remove(f)

    print(f"Cleaned: {folder}")

clean_folder(output_image_dir)
clean_folder(output_mask_dir)

def random_crop_50_75(img, mask):
    h, w = img.shape[:2]

    area_ratio = random.uniform(0.50, 0.75)

    # convert area ratio to dimension scale
    scale = np.sqrt(area_ratio)

    crop_h = int(h * scale)
    crop_w = int(w * scale)

    y = random.randint(0, h - crop_h)
    x = random.randint(0, w - crop_w)

    # crop image and mask
    img_crop = img[y:y + crop_h, x:x + crop_w]
    mask_crop = mask[y:y + crop_h, x:x + crop_w]

    return img_crop, mask_crop
```

```python
import shutil

num_crops_per_image = 5
image_dir = "/mnt/c/Neethu/starDist_training/ground_truth/train/images/"
mask_dir = "/mnt/c/Neethu/starDist_training/ground_truth/train/masks/"

image_paths = glob(os.path.join(image_dir, "*.tif")) + \
              glob(os.path.join(image_dir, "*.tiff"))

for img_path in image_paths:

    filename = os.path.basename(img_path)
    name = os.path.splitext(filename)[0]

    # corresponding mask
    mask_path_tif = os.path.join(mask_dir, name + ".tif")
    mask_path_tiff = os.path.join(mask_dir, name + ".tiff")

    if os.path.exists(mask_path_tif):
        mask_path = mask_path_tif
    elif os.path.exists(mask_path_tiff):
        mask_path = mask_path_tiff
    else:
        print(f"Mask not found for {filename}")
        continue

    shutil.copy2(
        img_path,
        os.path.join(output_image_dir, filename)
    )

    shutil.copy2(
        mask_path,
        os.path.join(output_mask_dir, os.path.basename(mask_path))
    )

    img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
    mask = cv2.imread(mask_path, cv2.IMREAD_UNCHANGED)

    if img is None or mask is None:
        print(f"Skipping {filename}")
        continue

    for i in range(num_crops_per_image):

        img_crop, mask_crop = random_crop_50_75(img, mask)

        out_img = os.path.join(
            output_image_dir,
            f"{name}_crop_{i}.tif"
        )

        out_mask = os.path.join(
            output_mask_dir,
            f"{name}_crop_{i}.tif"
        )

        
        imwrite(
            out_img,
            img_crop,
            compression=None
        )

        imwrite(
            out_mask,
            mask_crop,
            compression=None
        )

    print(f"Processed: {filename}")
print("All augmentations completed.")
```

# Data Augmentation

You can define a function/callable that applies augmentation to each batch of the data generator.  
We here use an `augmenter` that applies random rotations, flips, and intensity changes, which are typically sensible for (2D) microscopy images (but you can disable augmentation by setting `augmenter = None`).

```python
def random_fliprot(img, mask): 
    assert img.ndim >= mask.ndim
    axes = tuple(range(mask.ndim))
    perm = tuple(np.random.permutation(axes))
    img = img.transpose(perm + tuple(range(mask.ndim, img.ndim))) 
    mask = mask.transpose(perm) 
    for ax in axes: 
        if np.random.rand() > 0.5:
            img = np.flip(img, axis=ax)
            mask = np.flip(mask, axis=ax)
    return img, mask 

def random_intensity_change(img):
    img = img*np.random.uniform(0.6,2) + np.random.uniform(-0.2,0.2)
    return img


def augmenter(x, y):
    """Augmentation of a single input/label image pair.
    x is an input image
    y is the corresponding ground-truth label image
    """
    x, y = random_fliprot(x, y)
    x = random_intensity_change(x)
    # add some gaussian noise
    sig = 0.02*np.random.uniform(0,1)
    x = x + sig*np.random.normal(0,1,x.shape)
    return x, y
```

# Evaluation and Detection Performance

Besides the losses and metrics during training, we can also quantitatively evaluate the actual detection/segmentation performance on the validation data by considering objects in the ground truth to be correctly matched if there are predicted objects with overlap (here [intersection over union (IoU)](https://en.wikipedia.org/wiki/Jaccard_index)) beyond a chosen IoU threshold $\tau$.

The corresponding matching statistics (average overlap, accuracy, recall, precision, etc.) are typically of greater practical relevance than the losses/metrics computed during training (but harder to formulate as a loss function). 
The value of $\tau$ can be between 0 (even slightly overlapping objects count as correctly predicted) and 1 (only pixel-perfectly overlapping objects count) and which $\tau$ to use depends on the needed segmentation precision/application.

Please see `help(matching)` for definitions of the abbreviations used in the evaluation below and see the Wikipedia page on [Sensitivity and specificity](https://en.wikipedia.org/wiki/Sensitivity_and_specificity) for further details.
